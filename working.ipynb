```python
def preprocess_tweet_text(tweet):
    # Remove urls
    tweet = re.sub(r"http\S+|www\S+|https\S+", '', tweet, flags=re.MULTILINE)
    # Remove user @ references and '#' from tweet
    tweet = re.sub('[^a-zA-Z]',' ', tweet)
    tweet.lower()
    tweet = [WordNetLemmatizer().lemmatize(word) for word in tweet.split(' ')]
    tweet = " ".join(tweet)
    # stopword removal
    tweet = [word for word in tweet.split(' ') if word not in set(stopwords.words('english'))]
    tweet=" ".join(tweet)
    # tweet = tweet.replace("\n","")
    
    return tweet
```

```python
def avg_sentence_length(text):
    sentences = nltk.sent_tokenize(text)
    total_words = sum(len(nltk.word_tokenize(sentence)) for sentence in sentences)
    avg_sent_length = total_words / len(sentences) if sentences else 0
    return int(avg_sent_length)
```

```python
df_div_tdb_block_inner['text'].apply(avg_sentence_length)
```

```python
def complex_words(text):
    cnt=0
    total_words=0
    words = nltk.word_tokenize(text)
    for word in words:
        total_words=total_words+1
        if len(word)>3:
            cnt=cnt+1
    return ((cnt/total_words)*100)
```

```python
df_div_tdb_block_inner['text'].apply(complex_words)
```

```python
def fog_index(text):
    avg_sent_length = avg_sentence_length(text)
    complex_percent = complex_words(text)
    fog_index_value = 0.4*(avg_sent_length+complex_percent)
    return fog_index_value
```

```python
df_div_tdb_block_inner['text'].apply(fog_index)
```

```python
def complex_words_count(text):
    cnt=0
    total_words=0
    words = nltk.word_tokenize(text)
    for word in words:
        total_words=total_words+1
        if len(word)>3:
            cnt=cnt+1
    return cnt
```
